```python
# Tìm tất cả các thẻ table với class chỉ định
tables = soup.find_all('table', class_="f1-table f1-table-with-data w-full")

# Kiểm tra nếu có bất kỳ thẻ table nào được tìm thấy
if tables:
    # Duyệt qua từng thẻ table
    for i, table in enumerate(tables):
        # In thẻ table với định dạng
        print(f"Table {i+1}:")
        print(table.prettify())
        print("---------------------------------------------------\n")
else:
    print("No tables found")
```

```python
import requests
from bs4 import BeautifulSoup

# Địa chỉ URL của trang xếp hạng tài xế
url = 'https://www.formula1.com/en/results/2024/drivers'

# Gửi yêu cầu HTTP đến URL
response = requests.get(url)

# Kiểm tra xem yêu cầu có thành công không
if response.status_code == 200:
    # Sử dụng BeautifulSoup để phân tích nội dung HTML
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Tìm bảng chứa dữ liệu xếp hạng
    # Thay đổi 'table-class-name' thành class chính xác của bảng dữ liệu
    table = soup.find('table', class_='resultsarchive-table')
    
    if table:
        # Lặp qua mỗi hàng trong bảng
        for row in table.find_all('tr')[1:]:  # bỏ qua hàng đầu tiên vì nó là tiêu đề
            columns = row.find_all('td')
            if columns:
                # Giả sử cột đầu tiên là tên tài xế, cột thứ hai là quốc tịch, v.v.
                driver_name = columns[1].text.strip()
                nationality = columns[2].text.strip()
                points = columns[3].text.strip()  # Giả sử điểm số nằm ở cột thứ ba
                print(f"{driver_name}, {nationality}, {points}")
    else:
        print("Không tìm thấy bảng.")
else:
    print("Failed to retrieve the website")
```
